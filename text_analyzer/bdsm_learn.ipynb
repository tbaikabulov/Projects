```python
import os
import requests
import shutil
import re
import pandas as pd
from random import randrange
from datetime import datetime
import numpy as np
```

```python
gay_codes = ['M/m', 'M/m+', 'M+/m', 'M+/m+', 'm-self', 'M/mf', 'MF/m', 'MF/mf']
```

```python
df = pd.read_csv('stories.csv')
```

```python
def str_to_list(s):
    return s.replace('[', '').replace(']', '').replace(',', '').replace("'", '').split()
```

```python
def interfere(list1, list2):
    return len(set(list1) & set(list2))>0
```

```python
df.shape
```

```python
df['gay'] = df.codes.apply(lambda x: interfere(str_to_list(x), gay_codes))
```

```python
df['link'] = df.story_id.apply(lambda x: f'https://www.bdsmlibrary.com/stories/story.php?storyid={x}')
```

```python
def filter_stories(df):
    data = df[df.rating >= 9]
    data = data[data.gay == 0]
    data = data[data.n_votes >=3]
    data = data[data.nc + data.slavery >=1]
    data = data.sort_values(by = 'rating', ascending = False)
    return data
```

```python
data = filter_stories(df)
```
